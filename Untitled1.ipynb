```python
import torch    # include 뜻
```

```python
# [] 리스트, {} 딕셔너리, () 튜플
X = torch.FloatTensor([[0,0],
                      [0,1],
                      [1,0],
                      [1,1]])
Y = torch.FloatTensor([[0], [1], [1], [0]])
```

```python
linear1 = torch.nn.Linear(2, 10, bias=True)     # linear -> linear1 변경
linear2 = torch.nn.Linear(10, 10, bias=True)    # linear2 추가
linear3 = torch.nn.Linear(10, 10, bias=True)    # linear3 추가
linear4 = torch.nn.Linear(10, 1, bias=True)     # linear4 추가
sigmoid = torch.nn.Sigmoid()
```

```python
# sigmoid 넣는 이유: 사이마다 단편화를 막기 위함
model = torch.nn.Sequential(linear1, sigmoid,
                            linear2, sigmoid,
                            linear3, sigmoid,
                            linear4, sigmoid)
```

```python
criterion = torch.nn.BCELoss    # BinaryCrossEntropy Loss -> 0, 1
optimizer = torch.optim.SGD(model.parameters(), lr=1)
```

```python
for step in range(10001):
    optimizer.zero_grad()
    hypothesis = model(X)
    cost = criterion(hypothesis, Y)
    cost.backward()
    optimizer.step()
    if step % 100 == 0:
        print(step, cost.item())
# linear2를 추가해서 XOR 연산이 0에 거의 수렴하는 결과를 얻게 됨
```

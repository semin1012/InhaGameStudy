```python
import torch    # include 뜻
```

```python
# [] 리스트, {} 딕셔너리, () 튜플
X = torch.FloatTensor([[0,0],
                      [0,1],
                      [1,0],
                      [1,1]])
Y = torch.FloatTensor([[0], [1], [1], [0]])
```

```python
linear = torch.nn.Linear(2, 1, bias=True)
sigmoid = torch.nn.Sigmoid()
```

```python
# linear -> sigmoid 차례대로 진행한다
model = torch.nn.Sequential(linear, sigmoid)
```

```python
# 현재 값과 정답 간의 차이를 비교하기 위해 사용한다.
criterion = torch.nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1)
```

```python
for step in range(10001):
    optimizer.zero_grad()   # 미분으로 학습한다. 초기화
    hypothesis = model(X)
    cost = criterion(hypothesis, Y) # 계산한다
    cost.backward()
    optimizer.step()        # 다음 스텝 네? 가져갑니다 아마도
    if step % 100 == 0:
        print(step, cost.item())
    # 근데 결괏값이 cost가 떨어지다가 더 안 떨어지게 된다 < 문제점
```
